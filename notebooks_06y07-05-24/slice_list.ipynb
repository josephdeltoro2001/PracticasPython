# List Slicing

```python
players = ['one', 'two', 'three', 'four', 'five']
print(players[1])
print(players)
print(type(players))
print(players[1:3])
```

```python
print(players[0:1])
print(players[:3])
```

```python
print(players[2:])
```

```python
print(players[:])
```

```python
print(players[-1])
print(players[:-2])
print(players[-2:])
```

```python
first_index = -5
print(players[first_index:])
```

```python
#pal√≠ndromo = oso, anitalavalatina
players.append('six')
players.append('seven')
players.append('eigth')
players.append('nine')
players.append('ten')
```

```python
print(players)
print(players[3:9:2])
print(players[::2])
print(players[1::2])
```

```python
numbers = list(range(1,11))
print(numbers)
print(len(numbers))

print(numbers[::-1])
```
