```python
# COPIA UNA VARIABLE
nombre = "Alexis"
nombre2 = nombre
nombre = "Eduardo"

print(nombre, nombre2)

numbers = ['one', 'two', 'three']
numbers_two = numbers

print(numbers, numbers_two, sep='\n')
numbers[0] = 'Alexis' # SOLO ESTOY MODIFICANDO LA PRIMERA LISTA

print(numbers, numbers_two, sep='\n') # QUE PASO
```

```python
numbers = ['one', 'two', 'three']
numbers_copy = numbers.copy()

numbers[0] = 'Alexis'
print(numbers, numbers_copy, sep='\n')
```

```python
numbers = ['one', 'two', 'three']
numbers_copy = numbers[:]

numbers[0] = 'Alexis'
print(numbers, numbers_copy, sep='\n')
```

```python
# CICLO
numbers_one = [1, 2, 3]
numbers_two = [4, 5, 6]

for number in numbers_two:
  numbers_one.append(number)

print(numbers_one)
```

```python
# Merge list
numbers_one = [1, 2, 3]
numbers_two = [4, 5, 6]

numbers = numbers_one + numbers_two
print(numbers)
```

```python
numbers_one = [1, 2, 3]
numbers_two = [4, 5, 6]

numbers_one.extend(numbers_two)
print(numbers_one)
```

```python
numbers_one = [1, 2, 3]
numbers_two = [4, 5, 6]

for number in (numbers_one, numbers_two):
  print(number)

for number in (numbers_one, numbers_two):
  for n in number:
    print(n)

numbers = []
for number in (numbers_one, numbers_two):
  for n in number:
    numbers.append(n)

print(numbers)
```

```python
# List comprehension
numbers = [n for number in (numbers_one, numbers_two) for n in number]
print(numbers)
```

```python
numbers_one = [1, 2, 3]
numbers_two = [4, 5, 6]

numbers = [*numbers_one, *numbers_two]
print(numbers)
```
